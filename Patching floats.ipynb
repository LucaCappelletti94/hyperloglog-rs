```python
with open("src/bias.rs", "r") as f:
    data = f.readlines()
```

```python
new_data = []

for line in data:
    original = line
    if "_" in line and line.count("_") == 1 and "." in line:
        number = line.strip().strip(",")
        number_of_mantissa = len(number.split(".")[1])
        if len(number) == number_of_mantissa + 1 + 4 + 1:
            new_number = "       {:.3f},\n".format(float(number))
            new_data.append(new_number)
            continue
        
    new_data.append(line)
print("".join(new_data))
```

```python
new_data = []

for line in data:
    original = line
    if "." in line and "       " in line and line.count(",") == 1 and "_" not in line:
        value = float(line.strip().strip(","))
        number_of_integer_digits = len(str(value).split(".")[0]) 
        
        number_of_floating_point_digits = 7 - number_of_integer_digits
        
        formatted_float = "{{:.{}f}}".format(
            number_of_floating_point_digits
        ).format(value)
        
        integer_portion, float_portion = formatted_float.split(".")
        
        float_portion = float_portion.rstrip("0")
        new_number = ""
        
        for i, digit in enumerate(reversed(integer_portion)):
            if "-" != digit and (i + 1) % 3 == 0:
                digit = "_" + digit
            new_number =  digit + new_number
                
        new_number = new_number.strip("_")
        formatted_float = new_number + "." +float_portion
        formatted_float = formatted_float.replace("__", "_")
        formatted_float = formatted_float.replace("-_", "-")
        
        if value < 0 and "-" not in formatted_float:
            formatted_float = "-" + formatted_float
        
        line = "       {},\n".format(formatted_float)
        
    new_data.append(line)
print("".join(new_data))
```
